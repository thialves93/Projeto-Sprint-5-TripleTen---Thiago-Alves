```python
import pandas as pd
import plotly.express as px

car_data = pd.read_csv('../vehicles_us.csv') # lendo os dados
print(car_data.info())
display(car_data.head())
```

```python
# Exemplo do projeto

fig = px.histogram(car_data, x="odometer") # criar um histograma
fig.show() # exibindo
print()
print()
fig = px.scatter(car_data, x="odometer", y="price") # criar um gráfico de dispersão
fig.show() # exibindo
```

```python
# Remover carros sem preço definido
car_data_clean = car_data.dropna(subset=['price'])
carros_baratos = car_data_clean.sort_values('price').head(10)
display(carros_baratos)

# Top 5 modelos mais baratos
top5_price = car_data.nsmallest(5, 'price')[['model', 'price']]
print(top5_price)
```

```python
# Valores únicos em ordem crescente
precos_ordenados = sorted(car_data['price'].unique())
print(precos_ordenados)
```

```python
# Quantos preços diferentes existem
quantidade_precos = car_data['price'].nunique()
print(f"Existem {quantidade_precos} preços únicos")
```

```python
# Os 10 preços que mais aparecem
precos_comuns = car_data['price'].value_counts().head(10)
print(precos_comuns)
```

```python
min_price = 1000  # ajuste de preço mínimo
common_prices_filtered = [p for p in precos_comuns.index if p is not None and p >= min_price]
cols = ['price','model_year','model','condition','odometer','transmission','type','paint_color']
dfp = car_data[car_data['price'].isin(common_prices_filtered)].copy()
dfp['odometer'] = pd.to_numeric(dfp['odometer'], errors='coerce')

# (opcional) evitar repetição do mesmo modelo no mesmo preço
dfp = dfp.sort_values(['price','model','odometer']).drop_duplicates(['price','model'], keep='first')

top3_per_price = (
    dfp
    .sort_values(['price','odometer','model_year'], ascending=[True,True,False])
    .groupby('price', group_keys=False)
    .head(3)[cols]
)

print(top3_per_price)
```
